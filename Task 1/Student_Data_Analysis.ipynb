```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_data.csv')
```

```python
gender_counts = df['gender'].value_counts()
print('No. of Students by gender')
print(gender_counts)
```

```python
avg_productivity = df.groupby("gender")['productivity_score'].mean()
print(avg_productivity)
```

```python
# average stress levels
avg_stress = df.groupby('gender')['stress_level'].mean()
print(avg_stress)
```

```python
# average phone usage
avg_phoneusage = df.groupby('gender')['phone_usage_hours'].mean()
print(avg_phoneusage)
```

```python
df.describe()
```

```python
# Visualization
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='gender', multiple='stack', kde=True, palette='viridis')
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
x = df['study_hours_per_day']
y = df['productivity_score']

plt.plot(x, label='study_hours_per_day')
plt.plot(y, label='productivity_score')
plt.xlabel('study_hours_per_day')
plt.ylabel('productivity_score')
```

```python
plt.scatter(x, y)
plt.xlabel('study hours per day')
plt.ylabel('productivity score')
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    data=df,
    x="age",
    hue="gender",
    fill=True,
    alpha=0.4
)

plt.title("KDE Plot: Age Distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
columns = ['student_id', 'age', 'study_hours_per_day', 'sleep_hours',
       'phone_usage_hours', 'social_media_hours', 'youtube_hours',
       'gaming_hours', 'breaks_per_day', 'coffee_intake_mg',
       'exercise_minutes', 'assignments_completed', 'attendance_percentage',
       'stress_level', 'focus_score', 'final_grade', 'productivity_score']

corr_matrix = df[columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
